# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame

```python
url = "https://presidentofindia.nic.in/former-presidents.htm"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
PresidentName = []
for item in soup.find_all("h3"):
    item = item.text
    item = item.strip()
    item = item.split("(")
    print(item)
    PresidentName.append(item[0])
```

```python
PresidentName
```

```python
Term =[]
for service in soup.find_all("div",class_="presidentListing"):
    #print(service.text)
    service = service.text
    service = service.split("\n")
    Term.append(service[2].replace("Term of Office: ",""))
```

```python
print(len(Term),len(PresidentName))
```

```python
import pandas as pd
df = pd.DataFrame({'Names':PresidentName,'Term of Office':Term})
```

```python
df
```

# Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’.
You have to scrape:

i) Top 10 ODI teams in men’s cricket along with the records for matches, points andrating.

ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

iii) Top 10 ODI bowlers along with the records of their team andrating.

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
soup=BeautifulSoup(page.content)
```

```python
page
```

```python
titles=soup.find_all('span',class_='u-hide-phablet')
tit=[]
for i in titles:
    tit.append(i.text.replace('\n',' '))
tit[:10]
```

```python
team_match=soup.find_all('td',class_='table-body__cell u-center-text')
team_matches=[]
team_points=[]
j=0
team_matches.append(soup.find('td',class_='rankings-block__banner--matches').text)
team_points.append(soup.find('td',class_='rankings-block__banner--points').text)
for i in team_match:
    if j==0 or j%2==0:
        team_matches.append(i.text.replace('\n',''))
    else:
        team_points.append(i.text.replace('\n',''))
    j+=1
print(team_matches[:10])
print(team_points[:10])
```

```python
rate=soup.find_all('td',class_='table-body__cell u-text-right rating')
team_rating=[]
team_rating.append(soup.find('td',class_='rankings-block__banner--rating u-text-right').text.replace('\n','').replace(' ',''))
for i in rate:
    team_rating.append(i.text.replace('\n','').replace(' ',''))
print(team_rating[:10])
```

```python
team=pd.DataFrame({})
team['Name']=tit[:10]
team['Matches']=team_matches[:10]
team['Points']=team_points[:10]
team['Rating']=team_rating[:10]
team
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame

i) Headline

ii) Time

iii) News Link

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
```

```python
page
```

```python
# get the headlines news and time into list
headlines =[]
time=[]

for h in soup.find_all("a",class_="LatestNews-headline"):
    h =h.text.strip()
    headlines.append(h)

for i in soup.find_all("time",class_="LatestNews-timestamp"):
    t = i.text.strip()
    time.append(t)
   
```

```python
# get the links of latest Headlines links
links = []
for item in soup.find_all("div",class_="LatestNews-container"):
    for l in item.find_all("a",class_="LatestNews-headline"):
        link = l.get("href")
        links.append(link)
    
```

```python
# check the data by print
print(links)
```

```python
print(len(headlines),len(time),len(links))
```

```python
# fit the list data in to Dataframe
import pandas as pd
CNBClatestNews = pd.DataFrame({"Headline":headlines,"Links":links,"Time":time})
```

```python
CNBClatestNews
```

# 6) Write a python program to scrape the details of most downloaded articles from AI in last 90

i) Paper Title

iii) Published Date

iv) Paper URL


```python
url = ("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
req = requests.get(url)
req
```

```python
soup = BeautifulSoup(req.content)
soup
```

```python
title=[]
authors=[]
publishdata=[]
paperurl=[]

for data in soup.find_all("h2",class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    data = data.text.strip()
    title.append(data)
    
for data in soup.find_all("span", class_="sc-1w3fpd7-0 dnCnAO"):
    data = data.text.strip()
    authors.append(data)

for data in soup.find_all("span",class_="sc-1thf9ly-2 dvggWt"):
    data = data.text.strip()
    publishdata.append(data)
```

```python
paperurl =[]
for item in soup.find_all("a",class_="sc-5smygv-0 fIXTHm"):
    l =item.get("href")
    paperurl.append(l)    
```

```python
print(paperurl)
```

```python
print(len(title),len(authors),len(publishdata),len(paperurl))
```

```python
import pandas as pd
AIarticles = pd.DataFrame({"Paper Title":title,"Authors":authors,"Published Date":publishdata,"Paper URL":paperurl})
```

# Scraping multiple Titles

```python
titles = [] # empty list for store the 

for i in Soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)

    
titles
```

```python
Prices = [] # empty list for store the 

for i in Soup.find_all('span',class_="double-line-ellipsis"):
    Prices.append(i.text.split('|')[0])

    
Prices
```

# Scraping multiple Locations

```python
location = [] # empty list

for i in Soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
images = []

for i in Soup.find_all("img",class_="no-img"):
    images.append(i.get('data-src'))
                       
images
```

```python
Ratings =[]

for i in Soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
                       
Ratings
```

```python
Cuisine = []

for i in Soup.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.split("|")[1].strip())
                       
Cuisine
```
